```python
###Predicting Leads which have a chance to convert
total_leads = marketing_qualified_leads.merge(closed_deals, how='left',left_on='mql_id',right_on='mql_id')
```

```python
total_leads['converted_flag'] = np.where(total_leads['won_date'].isna(),0,1)
```

```python
total_leads['converted_flag'].value_counts()
```

```python
###Out of a 8000, 842 leads have converted.Let's explore the data and look at some behaviour of converted leads.
converted_leads = total_leads[total_leads['converted_flag'] == 1]
converted_leads['origin'].value_counts()
```

```python
converted_leads['won_date'] = pd.to_datetime(converted_leads['won_date'])
converted_leads['first_contact_date'] = pd.to_datetime(converted_leads['first_contact_date'])
```

```python
###Looks like organic and paid search are the best lead scourcs for conversion.Not surprising though
##We can build a predictive model for lead conversion but wwe do not have any negative samples in the dataset for that.
##We will be identifying leads which have a higher chance of conversion base on the time to convert and SR assigned.
converted_leads['time_to_convert'] = (converted_leads['won_date'] - converted_leads['first_contact_date']).dt.days
```

```python
print('Avg time to convert {}'.format(converted_leads.time_to_convert.mean()))
print('Min time to convert {}'.format(converted_leads.time_to_convert.min()))
print('Max time to convert {}'.format(converted_leads.time_to_convert.max()))
```

```python
###It takes an average of 48 days for a lead to convert. We can look at the outliers on this data and see which lead has gone beyond that
converted_leads = converted_leads[converted_leads['time_to_convert']>=0]
###We should also look at SR level i.e. how long is an SR taking to close the lead and his outliers
sr_convert = converted_leads.groupby('sr_id')['time_to_convert'].mean()
```

```python
###LTV calculations
###For converted seller, we take the transcations (price+shipping) for LTV calculations for the platform
###Getting the transactional data for the dealers
total_sales_seller = closed_deals.merge(order_items,how='left', left_on='seller_id',right_on='seller_id')
```

```python
total_sales_seller.head()
```

```python
###Assuming these are sales till today since won data.We will be calclating LTV on that basis
total_sales_seller['won_date'] = pd.to_datetime(total_sales_seller['won_date'])
total_sales_seller['till_date'] = pd.datetime.now()
```

```python
###Till_date_sales
total_sales_seller['no_of_days_till_date'] = total_sales_seller['till_date'] - total_sales_seller['won_date']
total_sales_seller['sales_price'] = total_sales_seller['price'] + total_sales_seller['freight_value']
agg_func = {'sales_price' : 'sum',
           'product_id':'count'}###To get total items
total_sales_grouped = total_sales_seller.groupby('seller_id').agg(agg_func).reset_index()
total_sales_grouped['average_purchase_value'] = total_sales_grouped['sales_price']/total_sales_grouped['product_id']
```

```python
###Total frequency rate can be determined by number of distinct orders of each sellers
###Calculating distinct orders for each seller
total_orders_unique = total_sales_seller.groupby(['seller_id','order_id']).size().reset_index().rename(columns={0:'count'})
total_orders_seller = total_orders_unique.groupby('seller_id')['order_id'].count().reset_index().rename(columns={'order_id':'total_orders'})
```

```python
###calculating frequency of orders. DIviding by number of days for total orders.
total_sales_grouped = total_sales_grouped.merge(total_orders_seller, how = 'left',left_on='seller_id',right_on='seller_id')
total_sales_seller = total_sales_seller[['seller_id','no_of_days_till_date']].groupby(['seller_id','no_of_days_till_date']).size().reset_index()
total_sales_grouped = total_sales_grouped.merge(total_sales_seller,
                                               how = 'left',left_on='seller_id',right_on='seller_id')
```

```python
total_sales_grouped.columns
```

```python
###Calculating frequency 
oneday = pd.Timedelta(days=1)
total_sales_grouped['frequency_rate'] = total_sales_grouped['total_orders']/(total_sales_grouped['no_of_days_till_date']/oneday)
```

```python
###Assuming an e-commerce seller stays in business for 5 years
total_sales_grouped['LTV'] = total_sales_grouped['average_purchase_value']*total_sales_grouped['frequency_rate']*365*5
```

```python
total_sales_grouped[['seller_id','LTV']].to_csv('LTV_values')
```
